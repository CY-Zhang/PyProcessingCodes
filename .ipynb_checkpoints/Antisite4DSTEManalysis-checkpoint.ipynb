### Test COM change for different combination of defect depth and sample thickness

```python
path = '/srv/home/chenyu/antisite/PRISMATIC/'
source_size = 90
for depth in range(3,26):
    for slice in range(0,depth):

        datacube = np.load(path+'Depth'+str(depth)+'/depth'+str(depth)+'_slice'+str(slice)+'_'+str(source_size)+'pmss.npy')

        savingName = 'COM_depth' + str(depth) + '_slice' + str(slice) + '_' + str(source_size) + 'pmss.npy'
        # generate meshgrid for weights
        kx,ky = datacube.shape[2:4]
        kx = np.linspace(0,kx-1,kx)
        ky = np.linspace(0,ky-1,ky)
        weights_x, weights_y = np.meshgrid(kx,ky)

        rx,ry = datacube.shape[0:2]
        COM = np.zeros((rx,ry,2))

        for ix in range(datacube.shape[0]):
            for iy in range(datacube.shape[1]):
                frame = datacube[ix,iy,:,:]
                COM[ix,iy,0] = np.average(weights_x,weights = frame)
                COM[ix,iy,1] = np.average(weights_y,weights = frame)

        np.save(path + 'Depth' + str(depth) + '/' + savingName,COM)
```

### Test different ways of generate integrated intensity to get some obvious difference

### Construct image from certain annular range
Find a combination of annular range and real space position that will give a high enough difference from antisite defect

#### Generate annular intensity view for each pixel in real space

```python
def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)
```

```python
path = '/srv/home/chenyu/antisite/PRISMATIC/'
source_size = 90
depth = 3
slice = 0
datacube = np.load(path+'Depth'+str(depth)+'/depth'+str(depth)+'_slice'+str(slice)+'_'+str(source_size)+'pmss.npy')

kx = np.linspace(-datacube.shape[2]/2,datacube.shape[2]/2,datacube.shape[2])
ky = np.linspace(-datacube.shape[3]/2,datacube.shape[3]/2,datacube.shape[3])
kx2,ky2 = np.meshgrid(kx,ky)
distance_from_center = np.sqrt(kx2**2 + ky2**2)
mask_stack = np.zeros((datacube.shape[2],datacube.shape[3],math.floor(distance_from_center.max())))

for dist in range(math.floor(distance_from_center.max())):
    mask_stack[:,:,dist] = (dist <= distance_from_center) & (distance_from_center <= dist+1)
    mask_stack[:,:,dist] = np.sign(mask_stack[:,:,dist])
```

```python
for depth in range(3,26):
    for slice in range(0,depth):

        datacube = np.load(path+'Depth'+str(depth)+'/depth'+str(depth)+'_slice'+str(slice)+'_'+str(source_size)+'pmss.npy')
        savingName = 'COM_depth' + str(depth) + '_slice' + str(slice) + '_' + str(source_size) + 'AnnularInt.npy'

        AnnularInt = np.zeros((datacube.shape[0],datacube.shape[1],mask_stack.shape[2]))

        for ix in range(AnnularInt.shape[0]):
            for iy in range(AnnularInt.shape[1]):
                frame = datacube[ix,iy,:,:]
                for dist in range(AnnularInt.shape[2]):
                    AnnularInt[ix,iy,dist] = np.sum(frame*mask_stack[:,:,dist])/np.sum(mask_stack[:,:,dist])

        np.save(path + 'Depth' + str(depth) + '/' + savingName, AnnularInt)
        print('Depth' + str(depth) + ' slice ' + str(slice) + ' finished.')
```

## Simulation with samller convergence angle

```python
folder = '/srv/home/chenyu/antisite/PRISMATIC_smallConv/Depth0/'
# simulation with 3.5mrad convergence angle
for slc in range(26):
    base_name = 'depth0_slice'+str(slc)
    base_ext = '_FPavg.npy'
    array_size = (41,41)

    imsize = (np.load(folder+base_name+'_X0_Y0'+base_ext)).shape
    out_sz = array_size + imsize
    output = np.zeros(out_sz,dtype=np.float32)
    for x in range(array_size[0]):
            for y in range(array_size[1]):
                output[x,y,:,:] = np.load(folder+base_name+'_X'+str(x)+'_Y'+str(y)+base_ext)
    output = np.squeeze(output)
    np.save( folder + 'Depth0_slice' + str(slc) + '.npy', output)
```
