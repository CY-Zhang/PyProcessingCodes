# Test difference between PRISMATIC 2D, PRISMATIC 3D, and EJK Multislice

This script aims to help solve the problem in thickness prediction using PRISMATIC simulation, where thickness estimated using PRISMATIC-Multislice is too thin compared to PACBED estimated result. Previous work with EJK-Multislice under 77mm camera length matches well with PACBED, which means something might be wrong in PRISMATIC-Multislice simulation. Things to check include:

1. Match between 2D and 3D output. -> <font color='red'>2D and 3D output shows different intensity when annular range > 200mrad. This different will be small if integrated intensity of 2D output is large enough so that difference cancel with each other.
2. <font color='black'>Match between PRISMATIC-Multislice and EJK-Multislice on the same structure with same probe positions. -> <font color='red'>PRISMATIC simulation result has much higher intensity on HAADF output ranges from 87.9 - 439.5 mrad, which can explain why thickness predicted by P-HAADF is too low. For now, P-HAADF result cannot be trusted, not sure how this will effect CBED/low angle intensities.
3. <font color='black'>Difference generated by parameters in PRISMATIC, mostly alpha-cutoff, potential boundary. See whether these can solve the descrepency between HAADF intensities.
4. Check difference between PACBEDs, see if CBED/PACBED/low angle intensities from PRISMATIC can be trusted.

## Test 1: 2D output match to 3D output

Do a PRISMATIC-Multislice simulation, generate 2D result and 3D result, compare 2D result to 3D with corresponding annular range. Intermediate thickness output was added by Luis and was not implemented in the first version.
Use exactly the same peak center position in 2D and 3D image, check whether integrated intensity match.

```python
# 2D output, collection angle ranges from 0 to 100 mrad, 16 phonons are used, output every 2 layers from start
path = '/srv/home/chenyu/antisite/STO_CL60/ConvergenceTest_015/'
nslice = 25
source_size = 90
integration_radius = 120
px_size = 17.5
intensity_2D = np.zeros(nslice)
low_angle = [0,50,100,150,200,250,300,350]
intensity_list = np.zeros((len(low_angle),nslice))

for iangle in range(len(low_angle)):

    for islice in range(nslice):

        name = '2Doutput_slice' + str(islice) + '_' + str(low_angle[iangle]) + '-' + str(low_angle[iangle]+50)
        file_2D = [image for image in os.listdir(path) if 'noPhonon' in image and name in image]
        file_2D = file_2D[0]
        image_2D = loadMRCfile(path + file_2D)

        if islice == 0:

            im = img_as_float(image_2D)
            
            if iangle == 0:
                im = np.amax(im) - im
                
            image_max = ndi.maximum_filter(im, size=5, mode='constant')
            coordinates = peak_local_max(im, min_distance=20)
            center = [coordinates[0,0], coordinates[0,1]]
            circular_mask = create_circular_mask(im.shape[0],im.shape[1],center, integration_radius/px_size)

        # Intensity is a list of averaged intensity inside the Sr mask
        intensity_list[iangle,islice] = (image_2D*circular_mask).sum()/circular_mask.sum()

# Display results
fig, axes = plt.subplots(1, 3, figsize=(8, 3), sharex=True, sharey=True)
ax = axes.ravel()
ax[0].imshow(image_2D, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(image_max, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('Maximum filter')

im2 = ax[2].imshow(image_2D, cmap=plt.cm.gray)
ax[2].autoscale(False)
ax[2].plot(coordinates[:, 1], coordinates[:, 0], 'r.')
ax[2].axis('off')
ax[2].set_title('Peak local max')
fig.colorbar(im2,ax=ax[2])

fig.tight_layout()

plt.show()

# print(intensity_2D)
```

```python
# 2D output, collection angle ranges from 112.9 mrad to 353.33 mrad
path = '/srv/home/chenyu/antisite/STO_CL60/ConvergenceTest_015/NoPhonon/'
nslice = 25
source_size = 90
integration_radius = 120
px_size = 17.5
# inner_angle = 300
# outer_angle = 350

intensity_3D = np.zeros(nslice)
# center = [ x-1 for x in center]
intensity_list_3D = np.zeros((len(low_angle),nslice))

for iangle in range(len(low_angle)):
    
    inner_angle = low_angle[iangle]
    outer_angle = inner_angle + 50
    print(inner_angle,outer_angle)

    for islice in range(nslice):

        name = 'slice' + str(islice) + '_'
        file_3D = [image for image in os.listdir(path) if 'P0001' in image and name in image and '2D' not in image]
        file_3D = file_3D[0]
#         print(path+file_3D)
        cube = loadMRCfile_3D(path + file_3D)

        image = np.sum(cube[:,inner_angle:outer_angle,:],axis=1)
        intensity_list_3D[iangle,islice] = (image*circular_mask).sum()/circular_mask.sum()
    
# Display results
fig, axes = plt.subplots(1, 3, figsize=(8, 3), sharex=True, sharey=True)
ax = axes.ravel()
ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(image_max, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('Maximum filter')

im2 = ax[2].imshow(image, cmap=plt.cm.gray)
ax[2].autoscale(False)
ax[2].plot(coordinates[:, 1], coordinates[:, 0], 'r.')
ax[2].axis('off')
ax[2].set_title('Peak local max')
fig.colorbar(im2,ax=ax[2])

fig.tight_layout()

plt.show()

# print(intensity_3D)
```

```python
import struct
import mrcfile
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.io
import time

def loadMRCfile(filepath):
    with open(filepath, mode='rb') as file: # b is important -> binary
        file.seek(0)
        read_bytes = file.read(8)
        frame_width = struct.unpack('<i',read_bytes[0:4])
        frame_width = frame_width[0]
        frame_height = struct.unpack('<i',read_bytes[4:8])
        frame_height = frame_height[0]
        image = np.zeros((frame_height,frame_width))
        file.seek(256*4)
        dataset = file.read(frame_width*frame_height*4)
        for irow in range(frame_height):
            for icol in range(frame_width):
                pixel = struct.unpack('<f',dataset[(irow*frame_width+icol)*4:(irow*frame_width+icol+1)*4])[0]
                image[irow,icol] = pixel
        return image
```

```python
def create_circular_mask(h, w, center=None, radius=None):

    if center is None: # use the middle of the image
        center = [int(w/2), int(h/2)]
    if radius is None: # use the smallest distance between the center and image walls
        radius = min(center[0], center[1], w-center[0], h-center[1])

    Y, X = np.ogrid[:h, :w]
    dist_from_center = np.sqrt((X - center[0])**2 + (Y-center[1])**2)

    mask = dist_from_center <= radius
    return mask
```

```python
import struct
import mrcfile
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.io
import time

def loadMRCfile_3D(filepath):
    with open(filepath, mode='rb') as file: # b is important -> binary
        file.seek(0)
        read_bytes = file.read(12)
        frame_width = struct.unpack('<i',read_bytes[0:4])
        frame_width = frame_width[0]
        frame_height = struct.unpack('<i',read_bytes[4:8])
        frame_height = frame_height[0]
        frame_layer = struct.unpack('<i',read_bytes[8:12])
        frame_layer = frame_layer[0]
        # frame_layer should be the slowest changing dimensiion in the data, thus the first layer in the reading loop
        image = np.zeros((frame_height,frame_width,frame_layer))

        file.seek(256*4)
        dataset = file.read(frame_width*frame_height*frame_layer*4)
        
        for ilayer in range(frame_layer):
            for irow in range(frame_height):
                for icol in range(frame_width):
                    pixel = struct.unpack('<f',dataset[(ilayer*frame_width*frame_height+irow*frame_width+icol)*4:(ilayer*frame_width*frame_height+irow*frame_width+icol+1)*4])[0]
                    image[irow,icol,ilayer] = pixel
        return image
```
