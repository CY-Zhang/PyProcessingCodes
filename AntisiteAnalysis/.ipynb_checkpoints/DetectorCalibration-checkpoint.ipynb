```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.io
import time
import glob,os
import cv2
from scipy import optimize
from skimage import feature
from pylab import *

from skimage import data, color
from skimage.transform import hough_circle, hough_circle_peaks
from skimage.feature import canny
from skimage.draw import circle_perimeter
from skimage.util import img_as_ubyte
```

This script serves to calibrate readout on each pixel from pixelated detector. The idea is the response/readout from pixelated detector may not/no longer be linear to inject beam current, in which case the readout needs to be calibrated before being compared to simulation PACBED.

```python
path = '/srv/home/chenyu/DEbackup/022619/'
file_list = glob.glob(path+'*.npy')
MeanInt_list = np.zeros(size(file_list))
BeamFraction_list = np.zeros(size(file_list))
for i in range(0,size(file_list)):
# for i in range(2,3):
    fullpath = file_list[i]
    ZeroBeam = np.load(fullpath)
    
    sample = ZeroBeam
    bw = 1.0*(sample > 5)
    selem = disk(5)
    bw2 = closing(bw, selem)
    edges = feature.canny(bw2)

    sample = sample*bw
    sample = sample.astype('uint8')
    temp = edges.astype('uint8')

    image = sample

    # Detect two radii
    hough_radii = np.arange(120,500,2)
    hough_res = hough_circle(temp, hough_radii)

    # Select the most prominent 5 circles
    accums, cx, cy, radii = hough_circle_peaks(hough_res, hough_radii,
                                               total_num_peaks=1)

    print(fullpath,cx,cy,radii)
      
    cx = int(np.average(cx))
    cy = int(np.average(cy))
    radii = int(np.average(radii))
    mask = np.heaviside(ZeroBeam-10,1)
    
    MeanInt_list[i] = (mask*ZeroBeam).sum()/np.count_nonzero(mask)
    
    if radii<240:    
        BeamFraction_list[i] = 1/np.count_nonzero(mask)
#     else:
#         BeamFraction_list[i] = 1/math.pi/radii/radii
        
    print(MeanInt_list[i])

#     fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10, 10))
#     print(cx,cy,radii)
#     circy, circx = circle_perimeter(cy,cx,radii)
    
#     ax.imshow(image, cmap=plt.cm.gray)
#     plt.scatter(circx, circy)
#     plt.show()
#     plt.imshow(temp)
```

```python
fig, axes = plt.subplots(1, 1, figsize=(10, 10))
axes.set_xlabel('Beam Fraction',fontsize=20)
axes.set_ylabel('Average Readout (counts)',fontsize=20)
# BeamFraction = 1/np.asarray(Radii_list)/np.asarray(Radii_list)
axes.scatter(BeamFraction_list,MeanInt_list,c='red')
# axes.scatter(BeamFraction_list,MeanInt_list*0.9,c='blue')
axes.set_xlim([0,np.amax(BeamFraction_list)])
axes.set_ylim([0,np.amax(MeanInt_list)])
np.save(path+'BeamFraction.npy',BeamFraction_list)
np.save(path+'MeanInt.npy',MeanInt_list)

# curve fitting results: Intensity counts = 1.383*10^7 * BeamFraction - 13.93
```
