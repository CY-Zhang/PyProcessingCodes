## Script used to process the Fe-bcc simulation result from Li

```python
import os
import scipy.io as sio
import scipy.misc as smisc
from scipy import signal
import numpy as np
import re as regexp
import math
import matplotlib.pyplot as plt
import glob
import scipy
```

```python
path = '/srv/home/chenyu/Fe_simulation/scratch/Fe_bcc/'
PACBED_stack = np.zeros((272,272,101))
for ilayer in range(101):
    subpath = path + '*slice' + str(ilayer) + '*'
    filelist = glob.glob(subpath)
    for ifile in range(len(filelist)):
        sample = np.load(filelist[ifile])
        PACBED_stack[:,:,ilayer] = PACBED_stack[:,:,ilayer] + np.squeeze(sample)
```
