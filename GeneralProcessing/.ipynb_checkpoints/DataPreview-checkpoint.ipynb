### Calculate diffraction disk centers

```python
ExpPath = '/srv/home/chenyu/DEbackup/101618/S1/'
ExpName = 'S1'

rx = 150
ry = 150

kx = 512
ky = 512

kx2,ky2 = np.meshgrid(np.linspace(0,kx-1,kx), np.linspace(0,ky-1,ky))
distance_from_center = np.sqrt((kx2 - 256)**2 + (ky2-256)**2)
mask = distance_from_center <= 52    #52px is the radius determined by drawing quadrant around zero disk
mask = mask* distance_from_center
mask_FT = np.fft.fft2(mask)
pos = np.zeros((rx,ry,2))

for ix in range(1,rx):
    for iy in range(ry):
#         print(ix,iy)
        # skip fitting for the first few frames
#         if ix == 0 and iy<5:
#             pos[ix,iy,0] = 263
#             pos[ix,iy,1] = 266
#             continue
        filenum = ix*rx+iy+1
        frame = np.load(ExpPath+'npy/'+ExpName+'_'+format(filenum,'05')+'.npy')
        frame = frame/frame.max()
        frame = np.fft.fftshift(np.fft.ifft2(np.fft.fft2(frame) * mask_FT)).real
        frame = frame[200:350,200:350]
        guess = (0,np.amax(frame),75,75,50,50)
        bounds = ([0,0,0,0,0,0],[np.inf,np.inf,150,150,np.inf, np.inf])
        # use new Gaussian fitting method with boundaries defined
        params = fitGaussianx2(frame, guess, bounds)
        # 200px offset when selecting the window
        pos[ix,iy,0] = params[2]+200 # xpos
        pos[ix,iy,1] = params[3]+200 # ypos
        
    if ix % 10 == 0:
        print("Finished line %d " % (ix))

np.save(ExpPath+'DiffCenter.npy',pos)
```

### Calculate COMs based on diffraction centers

```python
ExpPath = '/srv/home/chenyu/DEbackup/101618/S3/'
ExpName = 'S3'
center = np.load(ExpPath+'DiffCenter.npy')

rx = 150
ry = 150

kx = 512
ky = 512

kx2,ky2 = np.meshgrid(np.linspace(0,kx-1,kx), np.linspace(0,ky-1,ky))
distance_from_center = np.sqrt((kx2 - 256)**2 + (ky2-256)**2)
mask = distance_from_center <= 52    #52px is the radius determined by drawing quadrant around zero disk
mask = mask* distance_from_center
mask_FT = np.fft.fft2(mask)
COM = np.zeros((rx,ry,2))

for ix in range(1,rx):
    for iy in range(ry):
        center_x = int(center[ix,iy,0])
        center_y = int(center[ix,iy,1])
        
        filenum = ix*rx+iy+1
        frame = np.load(ExpPath+'npy/'+ExpName+'_'+format(filenum,'05')+'.npy')
        frame = frame/frame.max()
        frame = frame[center_y-75:center_y+75,center_x-75:center_x+75]

        kx = np.linspace(0,149,150)
        ky = np.linspace(0,149,150)
        weights_x, weights_y = np.meshgrid(kx,ky)

        COM[ix,iy,0] = np.average(weights_x,weights = frame)
        COM[ix,iy,1] = np.average(weights_y,weights = frame)
        
    if ix % 10 == 0:
        print("Finished line %d " % (ix))

np.save(ExpPath+'ZeorDisk_COM.npy',COM)
```

```python
# New Gaussian fitting method with boundary and initial guess

def gaussianx2(xdata_tuple,bg,height, center_x, center_y, width_x, width_y):

    (x, y) = xdata_tuple
    width_x = float(width_x)
    width_y = float(width_y)
    g = height*exp(-(((center_x-x)/width_x)**2+((center_y-y)/width_y)**2)/2)+bg
    return g.ravel()

def fitGaussianx2(data, guess, bounds):
    x = np.linspace(0, data.shape[1]-1, data.shape[1])
    y = np.linspace(0, data.shape[0]-1, data.shape[0])
    x, y = np.meshgrid(x, y)
    xdata_tuple = (x,y)
    popt, pcov = optimize.curve_fit(gaussianx2, xdata_tuple, data.ravel(), p0=guess,bounds=bounds,method='trf',verbose=0,maxfev=100000)
    return popt
```

```python
def gaussian(height, center_x, center_y, width_x, width_y):

    width_x = float(width_x)
    width_y = float(width_y)
    return lambda x,y: height*exp(
                 -(((center_x-x)/width_x)**2+((center_y-y)/width_y)**2)/2)
 
def moments(data):

    total = data.sum()
    X, Y = indices(data.shape)
    x = (X*data).sum()/total
    y = (Y*data).sum()/total
    col = data[:, int(y)]
    width_x = sqrt(abs((arange(col.size)-y)**2*col).sum()/col.sum())
    row = data[int(x), :]
    width_y = sqrt(abs((arange(row.size)-x)**2*row).sum()/row.sum())
    height = data.max()
    return height, x, y, width_x, width_y
 
def fitgaussian(data):

    params = moments(data)
    errorfunction = lambda p: ravel(gaussian(*p)(*indices(data.shape)) -
                                  data)
    p, success = optimize.leastsq(errorfunction, params)
    return p
```
