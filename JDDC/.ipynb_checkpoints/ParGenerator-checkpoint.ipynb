# Script to generate parameter files for global parameter optimization

```python
import os
import scipy.io as sio
import scipy.misc as smisc
import numpy as np
import re as regexp
import math
import matplotlib.pyplot as plt
import scipy
import glob,os
```

```python
# Selections for all four parameters
RatioRange = [0.3]
RegYRange = [1,10,100,500]
Y1L2Range = [1,10,100,500]
Y2L2Range = [50,75,100]
total_jobs = 4
reps = 10
```

```python
# Create folders
base = 'D:/2020/JDDC/GlobalParSearch_2frames/'
for i1 in range(len(RatioRange)):
    path = str(RatioRange[i1]) + '/'
    os.mkdir(base + path)
    for i2 in range(len(RegYRange)):
        path = str(RatioRange[i1]) + '/' + str(RegYRange[i2]) + '/'
        os.mkdir(base + path)
        for i3 in range(len(Y1L2Range)):
            path = str(RatioRange[i1]) + '/' + str(RegYRange[i2]) + '/' + str(Y1L2Range[i3]) + '/'
            os.mkdir(base + path)
            for i4 in range(len(Y2L2Range)):
                path = str(RatioRange[i1]) + '/' + str(RegYRange[i2]) + '/' + str(Y1L2Range[i3]) + '/' + str(Y2L2Range[i4]) + '/'
                os.mkdir(base + path)
```

```python
# Create parameter file for single batch
default_file = 'D:/2020/JDDC/GlobalParSearch_2frames/2Frames_B1.par'
with open(default_file) as f:
    datafile = f.readlines()
    for i1 in range(len(RatioRange)):
        for i2 in range(len(RegYRange)):
            for i3 in range(len(Y1L2Range)):
                for i4 in range(len(Y2L2Range)):
                    for i5 in range(reps):
                        path = str(RatioRange[i1]) + '/' + str(RegYRange[i2]) + '/' + str(Y1L2Range[i3]) + '/' + str(Y2L2Range[i4]) + '/'
                        par_file = base + path + '10Frames_B'+str(i5+1)+'.par'
                        print(par_file)
                        f2 = open(par_file,"w+")
                        for i in range(len(datafile)):
    #                         print(datafile[i])
                            temp = datafile[i].split(' ')
                            if temp[0] == 'templateNumOffset':
                                f2.write('templateNumOffset ' + str(i5*10) + '\n')
                            elif  temp[0] == 'centerFileName':
                                f2.write('centerFileName /srv/home/chenyu/JDDC/Experiment/033018/0.5us/2Frames/atoms_'+str(i5+1)+'.txt\n')
                            elif temp[0] == 'weightRatioPenalty':
                                f2.write('weightRatioPenalty ' + str(RatioRange[i1]) + '\n')
                            elif temp[0] == 'weightRegY':
                                f2.write('weightRegY ' + str(RegYRange[i2]) + '\n')
                            elif temp[0] == 'weightY1L2':
                                f2.write('weightY1L2 ' + str(Y1L2Range[i3]) + '\n')
                            elif temp[0] == 'weightY2L2':
                                f2.write('weightY2L2 ' + str(Y2L2Range[i4]) + '\n')
                            elif temp[0] == 'saveDirectory':
                                f2.write('saveDirectory ' + path + '2Frames_B'+str(i5+1)+'\n')
                            else:
                                f2.write(datafile[i])
                        f2.close()
```

```python
# Create job submission files
base = 'D:/2020/JDDC/GlobalParSearch_2frames/'
njobs_per_submission = 50
ifile = 0
isubmission = 0
for i1 in range(len(RatioRange)):
    for i2 in range(len(RegYRange)):
        for i3 in range(len(Y1L2Range)):
            for i4 in range(len(Y2L2Range)):
                for i5 in range(reps):
                    path = str(RatioRange[i1]) + '/' + str(RegYRange[i2]) + '/' + str(Y1L2Range[i3]) + '/' + str(Y2L2Range[i4]) + '/'
                    par_file = base + path + '2Frames_B'+str(i5+1)+'.par'
                    if ifile == 0:
                        submission_file = base + 'Submission_' + str(isubmission) + '.sh'
                        print(submission_file)
                        f2 = open(submission_file,"w+")
                        f2.write('#!/bin/sh\n\n')
                        f2.write('#SBATCH --job-name=Test_' + str(isubmission) + '\n')
                        f2.write('#SBATCH --error=Test_' + str(isubmission) + '.err\n')
                        f2.write('#SBATCH --output=Test_' + str(isubmission) + '.out\n')
                        f2.write('#SBATCH --time=7-00:00:00\n')
                        f2.write('#SBATCH --nodes=1\n')
                        f2.write('#SBATCH --ntasks=20\n')
                        f2.write('##SBATCH --export=ALL\n')
                        f2.write('echo "Started on:"\ndate\n')
                        f2.write('export LD_LIBRARY_PATH="/usr/local/openblas/0.2.20/lib:/usr/local/suitesparse/5.2.0/lib64:/srv/home/chenyu/anaconda/lib:/srv/home/chenyu/bin/JDDC/quocGCC"\n\n')
                        f2.write('module load suitesparse/5.2.0\n')
                        f2.write('module load lapack/3.8.0\n')
                        f2.write('source activate\n')
                        f2.write('conda activate base\n')
                        isubmission = isubmission + 1
                    if i5 == 1:
                        continue
                    f2.write('/srv/home/chenyu/bin/JDDC/quocGCC/projects/electronMicroscopy/jointUndistDenoise ' + path + '10Frames_B'+str(i5+1)+'.par\n\n')
                    f2.write('mkdir ' + path + '2Frames_B'+str(i5+1)+'_results\n')
                    f2.write('mv ' + path + '2Frames_B'+str(i5+1)+'/*.csv ' + path + '2Frames_B'+str(i5+1)+'_results\n')
                    f2.write('mv ' + path + '2Frames_B'+str(i5+1)+'/*.txt ' + path + '2Frames_B'+str(i5+1)+'_results\n')
                    f2.write('mv ' + path + '2Frames_B'+str(i5+1)+'/bump-0_003-marked.png ' + path + '2Frames_B'+str(i5+1)+'_results\n')
                    f2.write('rm -r ' + path + '2Frames_B'+str(i5+1)+'/\n\n')
                    ifile = ifile + 1
                    if ifile == njobs_per_submission:
                        ifile = 0
                        f2.write('echo "Finished on:"\ndate\n')
                        f2.write('source deactivate\n')
                        f2.close()
                        # After the whole file writing, replace Windows style \r\n with Unix style \n
                        fileContents = open(submission_file,"r").read()
                        f2 = open(submission_file,"w", newline="\n")
                        f2.write(fileContents)
                        f2.close()
```
